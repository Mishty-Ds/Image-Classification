```python
#Fetch all the urls for images
Image_urls=[]
for i in Product_urls:
    for j in driver.find_elements_by_xpath('//img[@class="s-image"]'):
        source=(j.get_attribute('src'))
        if source is not None:
             if(source[0:4] == 'http'):
                    Image_urls.append(source)
                    
        
```
